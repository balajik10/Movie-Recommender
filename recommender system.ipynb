```python
import ast

genres_str = '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

genres_list = ast.literal_eval(genres_str)

print(genres_list)
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert5(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=5:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert5)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample data
descriptions = [
    "In the 22nd century, a paraplegic marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action adventure fantasy science fiction.",
    "Captain Barbossa, long believed to be dead, has come back to life and is headed to the edge of the earth with Will Turner and Elizabeth Swann. But nothing is quite as it seems. Adventure fantasy action."
]

# Query movie description
query_description = "A marine faces a unique mission on the moon Pandora and must choose between following orders and protecting an alien civilization. Adventure and conflict ensue."

# TF-IDF vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(descriptions)

# Compute cosine similarity between query and existing descriptions
cosine_sim_query = cosine_similarity(vectorizer.transform([query_description]), tfidf_matrix)[0]

# Set a similarity threshold (you can adjust this based on your preference)
similarity_threshold = 0.5

# Check if similarity is above the threshold
similar_movies = [index for index, sim in enumerate(cosine_sim_query) if sim > similarity_threshold]

# Print the result
if similar_movies:
    print(f"The movie description is similar to movies {similar_movies}.")
else:
    print("No similar movies found.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample data
description1 = "In the 22nd century, a paraplegic marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action adventure fantasy science fiction."

description2 = "Captain Barbossa, long believed to be dead, has come back to life and is headed to the edge of the earth with Will Turner and Elizabeth Swann. But nothing is quite as it seems. Adventure fantasy action."

# TF-IDF vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([description1, description2])

# Compute cosine similarity between the two descriptions
cosine_similarities = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1]).flatten()

# Set a similarity threshold (you can adjust this based on your preference)
similarity_threshold = 0.5

# Print the result
if cosine_similarities[0] > similarity_threshold:
    print("The movie descriptions are similar.")
else:
    print("The movie descriptions are not similar.")
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text): 
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]

    for i in movies_list:
        new_df.iloc[i[0]].title
        print(new_df.iloc[i[0]].title)

```
